```python
import pandas as pd

# If the CSV is in the same directory as your notebook:
#df = pd.read_csv('your_file_name.csv')

# If the CSV is in a different directory (provide the full path):
# df = pd.read_csv('C:/Users/YourUser/Documents/data/your_file_name.csv')
# Note: Use forward slashes (/) or raw strings (r'...') for paths in Windows to avoid escape sequence issues.

data = pd.read_csv('happydataset_Week4.csv')
data
```

```python
data.sort_values('avg_income', inplace=True)
#we said we will sort by avg_income
#inplace will change the dataset into a sorted one,
#otherwise, we will get a copy of the sorted dataset

richest = data[ data['avg_income'] > 15000 ]
# this is a filter
#richest is called a dataframe
 
richest.iloc[0]
#pulling out the first row
richest.iloc[-1]
#pulling out the last row
richest.iloc[0:5]
#pulling out a range from 1st to 6th


```

```python

#now i want to find the mean of the richest
import numpy as np
rich_mean = np.mean(richest['avg_income'])

all_mean = np.mean(data['avg_income'])
print(all_mean, rich_mean)
```

```python
plt.scatter(richest['avg_income'],richest['happyScore'])
plt.text(richest.iloc[0]['avg_income'],
         richest.iloc[0]['happyScore'],
        richest.iloc[0]['country'])
#we labelled the lowest country in richest group

#if we want to label London at (56,36), we can do
#plt.text(s='London',y=36,x=56)
#plt.text(y=36,x=56,s='London')
#plt.text(56,36,'London')
```

```python
#now we label all points on a scatterplot
#by using iterrows which is iterating over rows

plt.scatter(richest['avg_income'], richest['happyScore'])


for k,row in richest.iterrows(): #this is a forloop
    plt.text(row['avg_income'],
             row['happyScore'],
             row['country'])
    

    
```

```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('happydataset_Week4.csv')

plt.xlabel('Inequality')
plt.ylabel('happy Score')
plt.scatter(data.income_inequality, [data.happyScore])
#data.varname is for numbers

plt.scatter(data.income_inequality, [data.happyScore], s=data.income_inequality*10, alpha=0.25)
#s = is for increasing the size of the dots
```

```python
#we try k.means
from sklearn.cluster import KMeans
import numpy as np

happy = data['happyScore'] #for strings
#we can also pull out the columns with the commands:
#data.happyScore
#data["happyScore"] for strings
income = data['avg_income']

income_happy = np.column_stack((income,happy))
print(income_happy)

km_res = KMeans(n_clusters=2).fit(income_happy)
clusters = km_res.cluster_centers_

plt.scatter(income,happy)
plt.scatter(clusters[:,0],clusters[:,1], s=1000, alpha=0.25)
#[:,0] means first column
```
